```python
import re
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
f = open('./blg_stats_2022-23.html', encoding='utf-8')

page_bs = BeautifulSoup(f.read(), 'html.parser')
```

```python
df = pd.DataFrame()

is_header_found = False
col_names = []

for row in page_bs.find_all('tr'):
    row_values = []
    col_targets = ['td'] if is_header_found else ['td','th']
    for col in row.find_all(col_targets):
        if col.name == 'th' and not is_header_found:
            is_header_found == True
            col_names.append(col.text)
        elif col.name == 'td':
            row_values.append(col.text)
    df = pd.concat([df, pd.DataFrame([row_values])])
```

```python
df.columns = ['RANK','PLAYER','TEAM'] + col_names[17:45] + ['PLUSMINUS']
```

```python
df = df[pd.notna(df['RANK'])]
```

```python
df['PLAYER'] = df['PLAYER'].str.replace('\n','')
df['TEAM'] = df['TEAM'].str.replace('\n','')
```

```python
a = np.array([re.sub('[0-9]','', name).split('#') for name in df['PLAYER']])
df['PLAYER'] = a[:,0]
df['POSITION'] = a[:,1]
```

```python
a = np.array([min.split(':') for min in df['MIN']])
min = pd.to_numeric(a[:,0]) + (pd.to_numeric(a[:,1]) / 60)
df['MIN'] = list(map(lambda x: round(x, 1), min))
df['MINPG'] = list(map(lambda x: round(x, 1), (df['MIN'] / pd.to_numeric(df['G']))))
```

```python
df['RANK'] = df['RANK'].astype(int)
df.index = df['RANK']
df = df.drop('RANK', axis=1)

df['G'] = df['G'].astype(int)
df['GS'] = df['GS'].astype(int)

df['PPG'] = df['PPG'].astype(float)
df['FGMPG'] = df['FGMPG'].astype(float)
df['FGAPG'] = df['FGAPG'].astype(float)
df['2FGMPG'] = df['2FGMPG'].astype(float)
df['2FGAPG'] = df['2FGAPG'].astype(float)
df['3FGMPG'] = df['3FGMPG'].astype(float)
df['3FGAPG'] = df['3FGAPG'].astype(float)
df['FTMPG'] = df['FTMPG'].astype(float)
df['FTAPG'] = df['FTAPG'].astype(float)
df['ORPG'] = df['ORPG'].astype(float)
df['DRPG'] = df['DRPG'].astype(float)
df['RPG'] = df['RPG'].astype(float)
df['APG'] = df['APG'].astype(float)
df['TOPG'] = df['TOPG'].astype(float)
df['STPG'] = df['STPG'].astype(float)
df['BSPG'] = df['BSPG'].astype(float)
df['BSRPG'] = df['BSRPG'].astype(float)
df['FPG'] = df['FPG'].astype(float)
df['FDPG'] = df['FDPG'].astype(float)
df['EFFPG'] = df['EFFPG'].astype(float)
df['PLUSMINUS'] = df['PLUSMINUS'].astype(float)
```
